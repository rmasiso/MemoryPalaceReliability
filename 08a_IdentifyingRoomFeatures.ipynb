```python
import networkx as nx
import numpy as np

adj_noself = adj_mat.copy()
adj_noself[np.eye(adj_noself.shape[0],dtype=bool)] = 0

plt.figure()
plt.imshow(adj_noself)

graph = nx.from_numpy_array(adj_noself)

# the basics
num_nodes = graph.number_of_nodes()
num_edges = graph.number_of_edges()
density   = nx.density(graph)

print(f"Number of nodes: {num_nodes}")
print(f"Number of edges: {num_edges}")
print(f"Density: {density}")

# node properties
degrees           = dict(graph.degree())
degree_centrality = nx.degree_centrality(graph)
betweenness       = nx.betweenness_centrality(graph, normalized=True, weight=None) 
closeness         = nx.closeness_centrality(graph, distance=None) 
pagerank          = nx.pagerank(graph, alpha=0.85, weight=None)


print("Degrees: ", degrees) ; print(" ")
print("Centrality: ", degree_centrality) ; print(" ")
print("Betweenness Centrality:", betweenness) ; print(" ")
print("Closeness Centrality:", closeness) ; print(" ")
print("PageRank:", pagerank) ; print(" ")

try: 
    eigenvector = nx.eigenvector_centrality(graph, weight='weight', max_iter=500) 
    print("Eigenvector Centrality:", eigenvector); print(" ")
except nx.NetworkXError:
    print("Eigenvector Centrality computation failed to converge or graph might be problematic."); print(" ")


clustering_coefficients = nx.clustering(graph)  # Local clustering per node

# mean path length
if nx.is_connected(graph): 
    avg_path_length = nx.average_shortest_path_length(graph)

    # network properties
    global_clustering_coefficient = nx.transitivity(graph)

    print(f"mean shortest path length: {avg_path_length}"); print(" ")
    print(f"global clustering coeff: {global_clustering_coefficient}")
    
    
    
# draw graph
plt.figure(figsize=(10,10))
nx.draw(graph, with_labels=True, node_color='skyblue', node_size=1500, edge_color='gray', font_size=20)
plt.title("node tree")
plt.savefig("mempal.png", dpi=300, bbox_inches='tight', transparent=True)  # you can also use .pdf, .svg, etc.

plt.show()
```

# compare the features against each other

```python
property_list = [
    'occupied_box_volume', 
    'boxcoll_volume', 
    'ratio_occupied_boxcoll_volume', 
    "area", 
    "manual_object_count", 
    "num_corners",
    "view_outside",
    "degree",
    'betweenness',
    'closeness',
    'eigenvector',
    'pagerank'
    ]


df = pd.read_csv("room_features_processed.csv")

n_properties = len(property_list)
n_pairs      = n_properties * (n_properties - 1) // 2
ncols        = 6
nrows        = round(n_pairs/ncols) 

# create subplot
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols * 4, nrows * 4)) 
axes = axes.flatten() 

completed_corrs = []
corr_results = {}
counter = 0
for ci, col1 in enumerate(property_list):
    for cj, col2 in enumerate(property_list):
        
        if cj>ci:
            
            x = df[col1]
            y = df[col2]

            corr = stats.spearmanr(x, y)[0]
            corr_results[(col1, col2)] = corr

            axes[counter].scatter(x,y)
            axes[counter].set_title(f"{corr:.3f}", fontsize=20)
            axes[counter].set_xlabel(col1)
            axes[counter].set_ylabel(col2)

            completed_corrs.append([col1, col2])
            
            counter+=1
            
plt.tight_layout()
        
sorted_corrs = sorted(corr_results.items(), key=lambda item: abs(item[1]) if not np.isnan(item[1]) else -1, reverse=True)
for (c1, c2), rho_val in sorted_corrs:
    if np.isnan(rho_val):
        print(f"  {c1} vs {c2}: NaN")
    else:
        print(f"  {c1} vs {c2}: {rho_val:.3f}")
        
        

        
```

```python
#  function to safely get correlation, checking both key orders
def get_correlation(feat1, feat2, corr_dict):
    if feat1 == feat2:
        return 1.0 # correlation with self is 1
    key1 = (feat1, feat2)
    key2 = (feat2, feat1)
    if key1 in corr_dict:
        return corr_dict[key1]
    elif key2 in corr_dict:
        return corr_dict[key2]
    else:
        return np.nan

corr_thresh1 = .7
corr_thresh2 = 0.5
new_prop_list = []
for (col1, col2), current_corr in corr_results.items():

    # if col1 has a high correlation with another pair, check if col2 has also a high correlation with another pair
    # the one with most high correlations is discarded

    if abs(current_corr) > corr_thresh1:

        temp_col1_corrs = [] ; count1 = 0
        temp_col2_corrs = [] ; count2 = 0

        for other_col in property_list:
            if other_col != col1 and other_col !=col2:
                temp_corr = get_correlation(col1, other_col, corr_results)
                if abs(temp_corr) > corr_thresh2:
                    count1+=1
                    temp_col1_corrs.append(f"{other_col}({temp_corr:.2f})")


        for other_col in property_list:
            if other_col != col1 and other_col !=col2:
                temp_corr = get_correlation(col2, other_col, corr_results)
                if abs(temp_corr) > corr_thresh2:
                    count2+=1
                    temp_col2_corrs.append(f"{other_col}({temp_corr:.2f})")


        print(f"  *** Comparing '{col1}' vs '{col2}': ***")
        print()
        print(f"    '{col1}' has {count1} correlations > {corr_thresh2} with others: \n        {', '.join(temp_col1_corrs) if temp_col1_corrs else 'None'}")
        print(f"    '{col2}' has {count2} correlations > {corr_thresh2} with others: \n        {', '.join(temp_col2_corrs) if temp_col2_corrs else 'None'}")
        print()

#             for (col1b, col2b) in corr_results.keys():
#                 if col1 in col1b or col2b:
#                     temp_col1_corrs.append(corr_results[(col1,)])
#                 if col2 in col1b or cold2b:
#                     temp_col2_corrs.append((corr_results[(col2,next_col)])
```
