```python
def GetRoomsToRecall(recall_info,print_tsv=False):
    '''recall info is a dict with stimfile info'''
    if str(recall_info['stim_file'].tolist()[0]) != 'nan':
        stim_names = recall_info['stim_file'].tolist()[0] # convert to string
        stim_names = list(map(str.strip, stim_names.split(';'))) # 
        rooms_to_recall = [PschopyRoomLabel2TranscrptionLabels[room_name] for room_name in stim_names]
        ridx_to_recall = [RoomObjectFileNameKey[rooms_to_recall[i]] for i in range(len(rooms_to_recall))]
        if print_tsv:
            print(recall_info)
        return rooms_to_recall, list(np.array(ridx_to_recall)-1) #zero indexing here
    else:
        neg_return = [-1, -1, -1, -1, -1]
        return neg_return, neg_return
```

```python



## STATS:
# average time spent talking GR per GR
# average time spent talking FR
# time spent on each room or object
# 
# correct number of room-object associations
# adjacent travel vs non-adjacent travel
# 

## time spent per trial
# trial_time = np.full((nSubj, len(trials)), fill_type=np.nan)

recall_freq = {}

behav_confmat = {"GR": np.zeros((23,23,len(subject_ids))),"FR": np.zeros((23,23,len(subject_ids)))}

behav_recall = {"GR": {"rooms": np.zeros((23,len(subject_ids))),
                       "objects": np.zeros((23,len(subject_ids))),
                      "raw_objects": np.zeros((23, len(subject_ids)))},
                "FR": {"rooms": np.zeros((23,len(subject_ids))),
                       "objects": np.zeros((23,len(subject_ids))),
                      "raw_objects": np.zeros((23, len(subject_ids)))}
               }

count = -1
for i, subj in enumerate(subject_ids[:]):
    # for i in [23]:

#     subj = subject_ids[i] # get subj string

    print(i,subject_ids[i])

    ###
    ### FOR a TRIAL (FR or GR), CLASSIFY TIMEPOINTS
    ###

    for ti, trial in enumerate(trials[:]):
        count += 1
        
        # row index
        row_idx = (df.index[df['subj_idx'] == i] & df.index[df['trial'] == trial]).tolist()[0]


        if 'recall' in task:

            ## GET start_TR and end_TR
            cushion = 0 #int(round(TR*4)) #add 4 TRs of cushion to end_TR/ which is ~5s (peak of hrf)
            recall_info_TR = GetRecallEventInfo(MakeRecallEventMatrices(subj, task, trial, TR, False), event_types,cushion) #don't trust event_matrices from TR or lower res b/c events with less than a second disappear
            start_TR, end_TR, run_TRs = recall_info_TR['start'], recall_info_TR['end'], recall_info_TR['run_TRs']
            segment_TRs = end_TR - start_TR #we cut out segment from start to end
            
            # in seconds
            dtt = 1 #s
            recall_info = GetRecallEventInfo(MakeRecallEventMatrices(subj, task, trial, dtt, False), event_types,cushion) #don't trust event_matrices from TR or lower res b/c events with less than a second disappear
            start, end, run_TRs = recall_info_TR['start'], recall_info_TR['end'], recall_info_TR['run_TRs']

            room_events = recall_info['event_matrices']['room_events']
            object_events_bysubj = recall_info['event_matrices']['object_events'] #object indices match fixed room indeces 
            object_raw_events = recall_info['event_matrices']['object_raw_events'] #object indeces are fixed for all subjs
            object_events = recall_info['event_matrices']['object_events'] #object indices match fixed room indeces 

            #### GET ROOM-OBJECT RECALL CORRECTNESS
            ## GET ROOM / OBJECT INFORMATION
            
            # where there is a room or object entry in matrix
            room_entries = np.where(room_events==1)[1]
            object_entries = np.where(object_events==1)[1]
            object_raw_entries = np.where(object_raw_events==1)[1]

            # how much time spent in a room or talking about an object
            rooms_remembered,rooms_time = np.unique(room_entries, return_counts=True)
            objs_remembered, objects_time = np.unique(object_entries, return_counts=True)
            objs_raw_remembered, objects_raw_time = np.unique(object_raw_entries, return_counts=True)
            
            # create a dict with room/object as key and value is the time spent talking about that
            rooms_time = { k:v for (k,v) in zip(rooms_remembered, rooms_time)}  
            objects_time = { k:v for (k,v) in zip(objs_remembered, objects_time)}  
            objects_raw_time = { k:v for (k,v) in zip(objs_raw_remembered, objects_raw_time)}  

            # order of which rooms and objects were remembered
            # first_transitions,first_item_numbers,all_transitions,all_item_numbers
            rooms_remembered_ordered = GetTransitionsFromEventMat(room_events)[-1] # 20230724
            objs_remembered_ordered = GetTransitionsFromEventMat(object_events)[-1] # the -1
            objs_raw_remembered_ordered = GetTransitionsFromEventMat(object_raw_events)[-1]
            
            ### ROOMS that were CUED on screen for participant to recall objects and rooms in detail
            cue_rooms, cue_ridx = GetRoomsToRecall(recall_info) # get cued rooms

            ## THE RIGHT WAY TO DO OVERALL ACCURACY for FRs is to count the rooms they remembered and count the objects they remembered.
            all_remembered = ( len(np.unique(rooms_remembered))+len(np.unique(objs_remembered)) ) / 46
            
            cue_remembered = len(set(cue_ridx).intersection(set(objs_remembered)))/len(cue_ridx)
#             remembered = np.sum(rooms_remembered==objs_remembered)/np.max([len(rooms_remembered), len(objs_remembered)])
            print(subj,trial,cue_ridx, rooms_remembered_ordered, objs_remembered_ordered,cue_remembered)

#             print(subj,trial,rooms_remembered,objs_remembered)

            #####
            ##### recall tally
            #####
            recalled_paired_objects = np.array(list(set(cue_ridx).intersection(set(objs_remembered))))
            print("RECALLED OBJECTS $$$$$$$$$$ ", recalled_paired_objects)
            
            if trial != 'FR':    
                
                #behav_confmat['GR'][cue_ridx,]
                behav_recall['GR']['rooms'][cue_ridx,i] +=1
                behav_recall['GR']['objects'][recalled_paired_objects,i] +=1

                # need to re-order so that i tally using                
                raw_obj_idx = Convert_SourceItem_to_PairedItem_Order(subj,recalled_paired_objects, room_obj_assignments,fromItem='roomidx',toItem='objidx')
                behav_recall['GR']['raw_objects'][raw_obj_idx,i] +=1
                
            else:
                behav_recall['FR']['rooms'][np.unique(rooms_remembered),i] +=1
                behav_recall['FR']['objects'][np.unique(objs_remembered),i] +=1
                behav_recall['FR']['raw_objects'][np.unique(objs_raw_remembered_ordered),i] +=1
                


            ### UPDATE DATAFRAME
            df.cued_rooms.iloc[row_idx] = cue_ridx if trial!='FR' else np.nan
            df.rooms_recalled.iloc[row_idx] = rooms_remembered_ordered
            df.objects_recalled.iloc[row_idx] = objs_remembered_ordered
            df.objects_raw_recalled.iloc[row_idx] = objs_raw_remembered_ordered
            df.all_correct.iloc[row_idx] = all_remembered
            df.cued_correct.iloc[row_idx] = cue_remembered if trial!='FR' else np.nan
            df.trial_time.iloc[row_idx] = end - start 
#             df.time_rooms.astype(object)
            df.time_rooms.iloc[row_idx] = [rooms_time]
            df.time_objects.iloc[row_idx] = [objects_time]
            df.time_raw_objects.iloc[row_idx] = [objects_raw_time]
#             df.at[df.columns.get_loc("time_objects"),row_idx] = list(objects_time)



## save tally that counts rooms and objects remembered for each recall task type
## GRs count objects that were in presented path
## FRs count both rooms and objects; but are also held separately for rooms or objects
save_obj("recall_behavior_tally.pkl", behav_recall)

```

# across-participant mean time comparison (for GRs)

```python

from scipy.stats import ttest_1samp, zscore, t


# ####
# Bonferroni Correction
# ####
def my_bonferroni(pvals, n_tests=23):
    corrected_pvals = np.minimum(np.array(pvals) * n_tests, 1.0)
    significance_bool = corrected_pvals < 0.05
    return significance_bool, corrected_pvals


def cohen_d_onesample(x, mu=0.0):
    """
    Compute Cohen's d for a one-sample t-test.
    d = (mean - mu) / sd
    """
    x = np.asarray(x)
    m = np.nanmean(x)
    s = np.nanstd(x, ddof=1)
    return (m - mu) / (s + 1e-12)

def global_mean_test(data, labels, label_type='item', n_tests=None):
    """
    Performs one-sample t-tests of each column in `data` against the global mean,
    applies Bonferroni correction, and computes Cohen's d effect sizes.

    Parameters
    ----------
    data : np.ndarray
        Shape (n_subjects, n_items)
    labels : list
        Item names
    label_type : str
        Column label name ('room', 'object', etc.)
    n_tests : int, optional
        Number of comparisons (defaults to number of items)

    Returns
    -------
    pandas.DataFrame
        Columns: label, mean_time, std_time, z_score, raw_p,
        corrected_p, significant, cohen_d
    """
    if n_tests is None:
        n_tests = data.shape[1]
    
    means = data.mean(axis=0)
    stds  = data.std(axis=0, ddof=1)
    z_scores = zscore(means)
    global_mean = means.mean()

    tvals, pvals, ds = [], [], []
    for i in range(data.shape[1]):
        x = data[:, i]
        t, p = ttest_1samp(x, global_mean)
        tvals.append(t)
        pvals.append(p)
        ds.append(cohen_d_onesample(x, mu=global_mean))

    sig, pvals_corr = my_bonferroni(pvals, n_tests=n_tests)

    df = pd.DataFrame({
        label_type: labels,
        'mean_time': means,
        'std_time': stds,
        'z_score': z_scores,
        'raw_p': pvals,
        'tvals': tvals,
        'corrected_p': pvals_corr,
        'cohen_d': ds,
        'significant': sig
    }).sort_values(by='mean_time', ascending=False)

    return df



###############
###############
###############


# ####
# RUN ANALYSES
# ####

###############


room_data = subj_room_times[:, :-1, :].mean(1)  # mean over trials, shape: (n_subj, 23)
object_data = subj_obj_raw_times[:,:-1:].mean(1)


# Analyze Rooms
room_df = global_mean_test(room_data, room_names, label_type='room')

# Analyze Objects
object_df = global_mean_test(object_data, obj_names, label_type='object')

# ====================
# PLOTS
# ====================
plt.figure(figsize=(12, 5))
plt.bar(room_df['room'], room_df['mean_time'], yerr=room_df['std_time'])
plt.xticks(rotation=90)
plt.ylabel('Mean Time (s)')
plt.title('Average Time Spent per Room')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 5))
plt.bar(object_df['object'], object_df['mean_time'], yerr=object_df['std_time'])
plt.xticks(rotation=90)
plt.ylabel('Mean Time (s)')
plt.title('Average Time Spent per Object')
plt.tight_layout()
plt.show()

# ====================
# PRINT RESULTS
# ====================
print("Top Rooms by Mean Time (Significant ones marked):")
print(room_df[['room', 'mean_time', 'tvals', 'z_score', 'corrected_p','cohen_d', 'significant']].head(23))

print("\nTop Objects by Mean Time (Significant ones marked):")
print(object_df[['object', 'mean_time','tvals', 'z_score', 'corrected_p','cohen_d', 'significant']].head(23))
```

```python
jumps = np.zeros((nSubj))               # actual contiguity proportion
chance_jumps = np.zeros((nSubj))        # baseline using "unrecalled" denominator
chance_fixed_jumps = np.zeros((nSubj))  # baseline using fixed 23 denominator

for s, subj in enumerate(subject_ids):
    total_traversals = 0
    valid_traversals = 0
    chance_vals = []         # "unrecalled" baseline per transition
    chance_fixed_vals = []   # fixed 23 baseline per transition

    path = FRdf[FRdf.subj == subj].rooms_recalled.to_list()[0]
    
    for ni in range(len(path)-1):
        source = path[ni]     # current recalled room
        target = path[ni+1]   # next recalled room

        ### SKIP SELF-LOOPS COMPLETELY
        if source == target:
            continue  # Skip everything else for this step

        ### VALID TRAVERSAL CHECK
        #########################
        shortest_path = nx.shortest_path(g, source, target)
        if len(shortest_path) > 1:  # counts only real transitions
            total_traversals += 1
        if len(shortest_path) == 2:  # contiguous (adjacent) transition
            valid_traversals += 1

        ### CHANCE TRAVERSAL CHECK (unrecalled set)
        ##############################################################
        unrecalled = set(path[ni+1:])
        neighbors = set(nx.neighbors(g, source))
        contiguous_unrecalled = neighbors & unrecalled
        if len(unrecalled) > 0:
            chance_prob = len(contiguous_unrecalled) / len(unrecalled)
            chance_vals.append(chance_prob)

        ### CHANCE TRAVERSAL CHECK â€” FIXED 23 VERSION
        ##############################################
        chance_prob_fixed = len(neighbors) / 22
        chance_fixed_vals.append(chance_prob_fixed)
                    
    ## per-subject results
    jumps[s] = valid_traversals / total_traversals if total_traversals > 0 else np.nan
    chance_jumps[s] = np.mean(chance_vals) if chance_vals else np.nan
    chance_fixed_jumps[s] = np.mean(chance_fixed_vals) if chance_fixed_vals else np.nan
```

```python
## calculate if traversals were above chance

import numpy as np
from scipy import stats

# Differences for each baseline
diff_shrinking = jumps - chance_jumps
diff_fixed = jumps - chance_fixed_jumps

# effective n after omitting NaNs (pairs present)
n_shr = np.isfinite(diff_shrinking).sum()
n_fix = np.isfinite(diff_fixed).sum()

# paired t-tests
t_shrinking, p_shrinking = stats.ttest_rel(jumps, chance_jumps, nan_policy='omit')
t_fixed, p_fixed = stats.ttest_rel(jumps, chance_fixed_jumps, nan_policy='omit')

# Cohen's d_z for paired tests
d_shrinking = t_shrinking / np.sqrt(n_shr)
d_fixed     = t_fixed     / np.sqrt(n_fix)

# print(f"Shrinking baseline: t={t_shrinking:.3f}, p={p_shrinking:.5f}")
# print(f"Fixed 22 baseline: t={t_fixed:.3f}, p={p_fixed:.5f}")

print(f"Shrinking baseline: t({n_shr-1}) = {t_shrinking:.3f}, p = {p_shrinking:.3e}, d_z = {d_shrinking:.3f}")
print(f"Fixed-22 baseline:  t({n_fix-1}) = {t_fixed:.3f},     p = {p_fixed:.3e},     d_z = {d_fixed:.3f}")

```
